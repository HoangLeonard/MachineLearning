How to use regression for classification
LR: $$ y = f_{θ}(x) = \sum(θ_{i}x_{i}) $$
$$θ = (X'X)^{-1}X'y$$

y: real value → discrete value

using a thresshold h if x >= h → 1 else x < h → 0 (h: hyper parameter) but we are assuming that there have the order in classes but its not. (and how to pick a good enough thresshold? → selecting model problem)

multi-classification problem: 
* divide it into K binary problems.
* K thresshold

```julia
using DelimitedFiles   
A = readdlm("/home/minhhn/Documents/MachineLearning/Dataset/Forbes/forbes.txt")
```

```julia
y = A[:,2]
x = A[:,1]
```

```julia
N = length(y)
```

```julia
X = [ones(N) x]
```

```julia
train(X, y) = inv(X' * X) * X' *y
```

```julia
θ = train(X, y)
```

```julia
using Plots
plot(x, y, st=:scatter, xlabel = "Temperature", ylabel = "Pressure", legend=false)
plot!([x[1], x[N]], [y[1], y[N]])
```

```julia
predict(θ, xNew) = xNew * θ
```

```julia
y207 = predict(θ, [1 207])
```
